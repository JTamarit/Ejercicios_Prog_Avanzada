**1.** *(5 puntos)* Implementa una clase *ColaPrioridad* que permita el almacenamiento de objetos ordenados por prioridad. Esta cola de prioridad tiene la peculiaridad de que mantiene un rango de prioridades (numéricas de tipo entero) válidas. Es decir, al instanciar objetos de la clase *ColaPrioridad*, se debe especificar una prioridad mínima y una máxima, y ninguna prioridad fuera de este rango será válida. Podéis asumir que el rango de prioridades siempre será pequeño (es decir, que la diferencia entre la prioridad mínima y máxima nunca será muy elevada), y que este rango no puede modificarse una vez el objeto de tipo *ColaPrioridad* ha sido creado. Teniendo esto en cuenta, la clase *ColaPrioridad* permitirá:

   * Insertar un objeto con una prioridad dada. Si la prioridad indicada está fuera de rango, se lanzará un error. Ejemplo: inserción del objeto *'a'* con prioridad 3.
   * Insertar un objeto, sin especificar prioridad. Esta operación será valida únicamente para objetos numéricos de tipo entero. El valor del número será su propia prioridad. Ejemplo: insertar el objeto 4 (cuya prioridad será 4 de manera implícita). Si la prioridad está fuera de rango, se lanzará un error.
   * Extraer el objeto de mayor prioridad en toda la cola de prioridad. En caso de empate (es decir, si más de un objeto tiene la prioridad mayor), el orden de extración será en base al orden de inserción (first in, first out). Es decir, para cada posible prioridad, se debe mantener una cola de objetos.
   * Extraer el objeto más prioritario dada una prioridad concreta. Si existe más de un objeto con la misma prioridad, el orden de extración será también en base al orden de inserción (first in, first out).
   * Funcionalidades extra.
      * En este ejercicio, de deben añadir a la clase *ColaPrioridad* más funcionalidades de libre elección. Ejemplo: un método para vaciar la cola.
      * Podéis también definir vuestras propias excepciones, heredando de la clase *Exception* proporcionada por Python. Ejemplo: implementar clase *ExcepcionPrioridadFueraDeRango*.

```python
import heapq
from random import randrange

class ColaPrioridad:
    
    '''Permite el almacenamiento de objetos ordenados por su prioridad.
    La prioridad más alta corresponde al valor más bajo de una lista númerica
    o del rango de prioridades en una tupla (prioridad,objeto)'''
    
    def __init__(self):
        self.lista_con_prioridad = []
        self.lista_numerica = []
        self.heap=[]
        self.__rango_inferior=0
        self.__rango_superior="" 
    
    def __introducir_rango_prioridad(self):
        n=0
        while n < 2:
            try:
                if n == 0:
                    self.__rango_superior = int(input("Rango Superior >>"))
                if n == 1:
                    self.__rango_inferior = int(input("Rango Inferior >>"))
                    
                if self.__rango_superior > 10:
                    raise RangoSuperiorIncorrecto('El rango superior deber ser menor de 10')
                if n==0 and self.__rango_superior <= 0:
                    raise RangoNegativo('El valor del rango superior  ha de ser mayor de 0')
                
                if self.__rango_inferior >= self.__rango_superior:
                    raise RangoInferiorIncorrecto('El rango inferior ha de ser menor que el rango superior')
                if n==1 and self.__rango_inferior < 0:
                    raise RangoNegativo('El valor del rango inferior no puede ser negativo')
                n += 1
                
            except ValueError:
                print("El valor ha de ser un entero")
            except RangoSuperiorIncorrecto as e:
                print(e)
            except RangoNegativo as e:
                print(e)
            except RangoInferiorIncorrecto as e:
                print(e)
       
        self.rango=[self.__rango_inferior, self.__rango_superior]
        self.rango=tuple(self.rango)
       
        return self.rango
    
    def crear_cola_prioridad(self,lista):
        tupla=()
        self.heap = list(lista)
        heapq.heapify(self.heap)
        
        if type(self.heap[0]) == type(tupla):
            
            self.__introducir_rango_prioridad()
            
        return self.heap
    
    def insertar_objeto(self,item):
        
        '''Insertar el objeto item en la cola de prioridad, si es una tupla(proridad,objeto) comprueba
        la prioridad esté dentro del rango'''
        
        tupla=()
        entero=int()
        
        if type(item)==type(entero):
            heapq.heappush(self.heap,item)
            
        elif type(item)==type(tupla) and item[0] in range(self.__rango_inferior,self.__rango_superior):
            heapq.heappush(self.heap,item)
        else:
            
            print(f"La prioridad del objeto {item} esta fuera de rango:{self.rango}")
       
        return print(self.heap)
        
    def extraer_objeto_mayor_proridad(self):
        
        '''Extraer objeto de mayor prioridad de la cola, cumple First In -Fisrt Out
        en caso de empate'''
        
        return heapq.heappop(self.heap)
        
    def extraer_objeto_por_prioridad(self,prioridad):
        '''
        Extrae un objeto dada una prioridad en concreto, cumple First In -First Out
        en caso de empate
        '''
        lista_prioridad_seleccionada =[]
        for item in self.heap:
            if item[0]== prioridad:
                lista_prioridad_seleccionada.append(item)
                
        elementos_eliminar=lista_prioridad_seleccionada
        
        for item in elementos_eliminar:
            
            self.heap.remove(item)
                
        return f"Los elementos estraidos con prioridad: {prioridad} son: {lista_prioridad_seleccionada}"
    
    def vaciar_cola(self):
        
        '''Funcionalidad extra para vaciar cola de manera First In - First Out'''
        
        while self.heap != []:
            
            heapq.heappop(self.heap)
            
        return self.heap
    
    def n_elementos_menor_proridad(self,n):
        ''' Obtener los n elementos de menor prioridad, empezamos por el último elemento
        introducido'''
        return heapq.nlargest(n, self.heap)
        
    def n_elementos_mayor_prioridad(self,n):
        ''' Obtener los n elementos de mayor prioridad, se cumple FIFO'''
        return heapq.nsmallest(n, self.heap)
        


class ExcepcionPrioridadFueraDeRango(Exception):
    ''' Clase creada para la captura de excepciones propias'''
    pass
class RangoSuperiorIncorrecto(ExcepcionPrioridadFueraDeRango):
    '''Se produce esta excepción cuando el rango superior de prioridad es mayor de 10'''
    pass
class RangoNegativo(ExcepcionPrioridadFueraDeRango):
    '''Se produce esta excepción cuando cualquier valor del rango de prioridad es negativo'''
    pass
class RangoInferiorIncorrecto(ExcepcionPrioridadFueraDeRango):
    '''Se produce esta excepción cuando del rango inferior de prioridad es menor que 0'''
    pass


listado_numerico= [4,8,1,40,12,34,23,2,12]
listado_con_prioridades=[(4,'email'),(2,'llamada'),(2,'teams'),
                         (1,'meeting'),(3,'fax'),(1,'visita')]


   
cola_prioridad= ColaPrioridad()
print(cola_prioridad.crear_cola_prioridad(listado_numerico))
cola_prioridad.insertar_objeto(200)
print(cola_prioridad.extraer_objeto_mayor_proridad())
print(cola_prioridad.n_elementos_menor_proridad(5))
print(cola_prioridad.n_elementos_mayor_prioridad(5))
print(cola_prioridad.vaciar_cola())

print(cola_prioridad.crear_cola_prioridad(listado_con_prioridades))
cola_prioridad.insertar_objeto((11,'nueva'))
print(cola_prioridad.extraer_objeto_mayor_proridad())
print(cola_prioridad.extraer_objeto_por_prioridad(2))
print(cola_prioridad.n_elementos_menor_proridad(2))
print(cola_prioridad.n_elementos_mayor_prioridad(2))
print(cola_prioridad.vaciar_cola())
```

**2.** *(2 puntos)* Implementa un algoritmo de ordenación que utilice la cola de prioridad del ejercicio 1 para ordenar de manera decreciente una lista de números comprendidos entre el 0 y el 100. La lista puede tener duplicados. Como orientación, podéis consultar [Heapsort](https://es.wikipedia.org/wiki/Heapsort), un algoritmo de ordenación basado en colas de prioridad.

```python


class OrdenacionListaHeapsort(ColaPrioridad):
    
    ''' Aplica algoritmo Heapsort para una lista de números desordena del 0-100 a partir
    de la cola de prioridad creada con el metodo de la clase ColaPrioridad'''
    
    def __init__(self):
        self.lista_aleatoria=[]
        self.lista_aletoria_ordenada=[]
    
    def crear_lista_numeros(self):
        
        '''Crea lista desordenada de número de 0 al 100, pueden haber duplicados'''
        
        self.lista_aleatoria=[(randrange(101)) for x in range(101)]
        
        return self.lista_aleatoria
    
    def algoritmo_heapsort(self):
        
        '''Retorna una lista de números ordenada a partir de una lista de numeros
        desordanda y su cola de prioridad (heap)'''
        
        for i in range(len(self.lista_aleatoria)):
            
            self.lista_aleatoria[i] = heapq.heappop(self.heap)
            
        self.lista_aleatoria_ordenada =list(self.lista_aleatoria)
        
        return self.lista_aleatoria_ordenada


heapsort = OrdenacionListaHeapsort()
lista=heapsort.crear_lista_numeros()
print(f"Lista aleatoria:\n{heapsort.crear_cola_prioridad(lista)}")
print(f"Lista ordenada con el Algoritmo Heapsort:\n{heapsort.algoritmo_heapsort()}")
```
